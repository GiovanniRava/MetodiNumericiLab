- **7.** Valutare in aritmetica floating point l'espressione
$$A(x)=\frac{1}{x}-\frac{1}{(x+1)}$$  nel vettore   $x=[10^0, 10^1,... ,10^{20}]$ contenente potenze successive  di $10$,  Si calcoli per ogni elemento di x l'errore relativo considerando come valore esatto quello calcolato usando  l'espressione algebricamente equivalente
$$
B(x)=\frac{1}{x (x+1)}.
$$ Si visualizzi l'errore relativo in un grafico in scala logaritmica su $x$ e $y$

```python
x= np.logspace(0,20,21)
A_X = (1/x) - (1/(x+1))
print("A_X = ", A_X)

B_X = 1/(x*(x+1))
print("B_X =", B_X)

erroreRel = np.abs(A_X-B_X) / np.abs(B_X)

plt.figure(figsize= (8,6))
plt.loglog(x, erroreRel, marker= 'o', linestyle = '-', color = 'b')
plt.xlabel("x")
plt.ylabel("Errore relativo")
plt.title("Errore relativo di A(x) rispetto a B(x)")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```

- **8.** Scrivere uno script Matlab che fornisca le approssimazioni di $e=\exp(1)$ ottenute con la formula
$f(n)=(1/n+1)^n$ per $n=10^k$ e  $k=0, ..., 16$. Visualizzare: un grafico con i valori di $f(n)$ al variare di n ed un grafico con i valori dell'errore relativo (considerando come valore esatto il valore math.exp(1)).  Cosa si osserva? Commentare i risultati ottenuti.

```python


import math

# Definiamo i valori di n come potenze di 10 da 10^0 a 10^16
k_values = np.arange(0, 17)
n_values = 10 ** k_values

# Calcoliamo f(n)
f_n = (1 / n_values + 1) ** n_values

# Valore esatto di e
e_exact = math.exp(1)

# Calcoliamo l'errore relativo
errore_relativo = np.abs(f_n - e_exact) / e_exact

# Creiamo il grafico dei valori di f(n)
plt.figure(figsize=(8,6))
plt.plot(n_values, f_n, marker='o', linestyle='-', label="f(n)")
plt.axhline(e_exact, color='r', linestyle='--', label="exp(1)")
plt.xscale('log')
plt.xlabel("n")
plt.ylabel("f(n)")
plt.title("Approssimazione di e con f(n)")
plt.legend()
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()

# Creiamo il grafico dell'errore relativo
plt.figure(figsize=(8,6))
plt.loglog(n_values, errore_relativo, marker='o', linestyle='-', color='b')
plt.xlabel("n")
plt.ylabel("Errore relativo")
plt.title("Errore relativo dell'approssimazione di e")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
